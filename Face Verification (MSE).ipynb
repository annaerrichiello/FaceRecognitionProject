```
import cv2

# Load the images
from skimage import io

img = io.imread('https://images.vanityfair.it/wp-content/uploads/2021/02/08190901/Jennifer-Aniston-950x684.jpg')
```

```
imgg1 = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
```

```
test = io.imread('https://assets.afcdn.com/album/D20200207/phalbm25895000_w980h638c1.jpg')
imgg2 =cv2.cvtColor(test,cv2.COLOR_BGR2GRAY)
```

```
image = io.imread('https://www.cinematographe.it/wp-content/uploads/2020/12/jennifer-lawrence-1.jpg') 
imgg3 = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY) 
```

```
import numpy as np
def mse(imageA, imageB):
    err = np.sum((imageA.astype("float") - imageB.astype("float")) ** 2)
    err /= float(imageA.shape[0] * imageA.shape[1])
    return err
```

```
mse(np.resize(imgg1, (400,400)), np.resize(imgg2, (400,400)))
```

```
mse(np.resize(imgg1, (400,400)), np.resize(imgg3, (400,400)))
```

```
mse(np.resize(imgg2, (400,400)), np.resize(imgg3, (400,400)))
```

```
# the mse is lower between imgg1, imgg2. This implies that imgg1 is more similar to imgg2, rather than imgg3
```
